```python
import matplotlib.pyplot as plt

#sample data for energy consumption over 6 months (in Mwh)
months= ['Jan','Feb','Mar', 'Apr','May','Jun']
energy_consumption=[1200,1300,1100,1500,1400,1600]

#create a line plot
plt.plot(months, energy_consumption,marker='o', color='b',linestyle='--')

#add titles and lables
plt.title('Energy Consumption Over 6 Months')
plt.xlabel('Months')
plt.ylabel('Energy Consumption (Mwh)')

#display the plot
plt.show()
```

```python
#create a bar graph
plt.bar(months, energy_consumption, color='g')

#add titles and lables
plt.title('Energy Consumption Over 6 Months')
plt.xlabel('Months')
plt.ylabel('Energy Consumption (Mwh)')

#display the plot
plt.show()
```

```python
#pie chart
plt.pie(energy_consumption, labels=months, autopct='%1.1f%%')

#add titles and lables
plt.title('Energy Consumption Over 6 Months')
```

```python
#create scatter plot
plt.scatter(months, energy_consumption, color='r', marker='o')

#add titles and lables
plt.title('Energy Consumption Over 6 Months')
plt.xlabel('Months')
plt.ylabel('Energy Consumption (Mwh)')
```

```python
#create a histo gram
plt.hist(energy_consumption, bins=5, color='y', edgecolor='black')

#add titles and lables
plt.title('Energy Consumption Over 6 Months')
plt.xlabel('Energy Consumption (Mwh)')
plt.ylabel('Frequency')
```

```python
#double line chart
y1=[2,3,4.5]

y2=[1,1.5,5]

#function to plot
plt.plot(y1)
plt.plot(y2)

#function add a legend
plt.legend(["blue","green"],loc="lower right")

#function to show the plot
plt.show()
```

```python
#pair plot
import seaborn as sns
import  pandas as pd
import matplotlib.pyplot as plt

#creating a data frame for pair plot
data={ "Solar": [1200,1500,1300],
      "Wind": [3400,3500,3200],
       "Hydropower": [2900,3100,2800],
       "Biomass": [2500,2700,2400]}

df=pd.DataFrame(data)

#create a pair plot
sns.pairplot(df)

#show the plot
plt.show()
```

```python
#heat map: shows correlation betn diff types of energy consumption (imp)

#create a heatmap to show correlation matrix
correlation_matrix=df.corr()

#plot heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidth=0.5)

#add title
plt.title('Correlation btwn Energy Sources')
plt.show()
```

```python
#box plot : gives distribution of energy consumption by source
sns.boxplot(data=df)
plt.title('Distribution of Energy consumption by Sources')
plt.show()
```

```python
#violin plot: used when one continuous feature and 1 categorical feature
sns.violinplot( data=df)
plt.title('Distribution of Energy consumption by Sources')
plt.show()
```

```python
#regression plot:
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

#sample data for energy consumption and carbon emission
energy_value= [100,200,300,400,500]
carbon_emissions= [10,20,25,40,50]

#create a dataframe with energy consumption and carbon emission
df_reg=pd.DataFrame({ 'Energy Consumption': energy_value,
                     'Carbon Emissions': carbon_emissions})

#create a regression plot
sns.regplot(x=energy_value, y=carbon_emissions, data=df_reg)

#add title
plt.title('Relationship bten energy cons and carbon emission')
plt.show()
```

```python
#3d line plot
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

fig=plt.figure()
ax=plt.axes(projection ='3d')

#define all 3 axes
z=np.linspace(0,1,1000)
x=z*np.sin(25*z)
y=z*np.cos(25*z)

#plot the line
ax.plot3D(x,y,z,'green')
ax.set_title('3D Line Plot')
plt.show()
```
